```python
#CONVERT ATTRITION AND OVERTIME 'YES' TO 1 AND 'NO' TO 0 FOR CALCULATION

df3['Attrition_numeric']=df['Attrition'].map({'Yes':1, 'No':0})
df3['overtime_numeric']=df['Over_Time'].map({'Yes':1 , 'No':0})
```

```python
#ATTRITION RATE ANALYSIS BASED ON AGE GROUP

df3['Age_group']=pd.cut(df3['Age'],bins=[17,28,38,48],labels=['under-28','under-38','under-50'])
age_group=df3.groupby('Age_group',observed=False)['Attrition_numeric'].mean().reset_index()
age_group['Attrition_rate']=age_group['Attrition_numeric']*100
print(age_group)

#VISULAIZE THE ATTRITION RATE BASED ON AGE GROUP

plt.figure(figsize=(8,5))
sns.barplot(data=age_group ,x='Age_group', y='Attrition_rate')
plt.title('ATTRITION ANALYSIS BASED ON AGE GROUP',fontsize=14,fontweight='bold')
plt.xlabel('age group')
plt.ylabel('attrition rate')
plt.ylim(0,25)
plt.show()
```

```python
#ATTRITION RATE ANALYSE BASED ON BUSINESS TRAVEL AND TRAVEL DISTANCE FROM HOME
df['travel_distance'] = df['DistanceFromHome (KM)'].apply(lambda x: 'Long (>15KM)' if x > 15 else 'Short (<=15KM)')
travel_commute = df.groupby(['Business_Travel', 'travel_distance'])['Attrition_numeric'].mean().reset_index()
travel_commute['Attrition_Rate (%)'] = travel_commute['Attrition_numeric'] * 100

#VISUALIZE THE ATTRITION RATE OF BUSINESS TRAVEL AND TRAVEL DISTANCE FROM HOME
plt.figure(figsize=(10, 6))
sns.barplot(data=travel_commute, x='Business_Travel', y='Attrition_Rate (%)', hue='travel_distance', palette='flare')
plt.title('Attrition Rate by Business Travel & Distance from home', fontsize=14, fontweight='bold')
plt.xlabel('Business Travel Frequency', fontsize=12)
plt.ylabel('Attrition Rate (%)', fontsize=12)
plt.legend(title='Distance from home')
plt.ylim(0,30)
plt.show()
```
